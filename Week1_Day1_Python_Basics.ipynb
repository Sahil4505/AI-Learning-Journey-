```python
#greater of two numbers
num1=int(input("Enter first number "))
num2=int(input("Enter second number "))
if(num1>num2):
  print(num1,"is greater than ",num2)
elif(num1==num2):
    print(num1,"is equal to ",num2)
else:
    print(num2,"is greater than ",num2)
```

```python
#taking input as String,float,bool
name=str(input("What's your name "))
height=float(input("What's your height(in meter's) "))
print("Your name is ",name)
height_in_feet=height*3.28084
dec=height_in_feet%1
feet=int(height_in_feet)
inch=round(dec*12)
print("Your height is ",feet,"feet",inch,"inch")
```

```python
#print largest among 3 numbers
num1=int(input("Enter first number "))
num2=int(input("Enter second number "))
num3=int(input("Enter third number "))
if(num1>num2):
  if(num1>num3):
    print(num1,"is the largest number")
  else:
    print(num3,"is the largest number")
else:
  if(num2>num3):
    print(num2,"is the largest number")
  else:
    print(num3,"is the largest number")
```

```python
#check leap year
year=int(input("Enter a year "))
if((year%4==0 and year%100!=0)or year%400==0):
  print(year,"is a leap year")
else:
  print(year,"is not a leap year")
```
