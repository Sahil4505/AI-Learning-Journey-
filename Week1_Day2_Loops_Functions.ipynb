```python
#functions without return
def add(a,b):
  print(a+b)

add(5,7)
#function with return
def mul(a,b):
  return a*b
print(mul(5,7))
```

```python
#is even
def is_even(n):
  if(n%2==0):
    return True
  return False
print(is_even(97))
print(is_even(98))
```

```python
#Factorial
def fact(n):
  mul=1
  while n>=1:
    mul=mul*n
    n-=1
  return mul
print("Fact of 5 is",fact(5))
```

```python
def max_of_three(a,b,c,):
  if(a>b):
    if(a>c):
      return a
    else:
      return c
  else:
    if(b>c):
      return b
    else:
      return c
print("Max of 127,130,111 is",max_of_three(127,130,111))
```

```python
#is_prime
def is_prime(n):
  if(n==1):
    return False
  for i in range(2,n):
    if(n%i==0):
      return False
  return True
print("3 is Prime -",is_prime(3))
```

```python
#Fibonacci series
def fib(n):
  print(0)
  print(1)
  a=0
  b=1
  c=0
  for i in range(2,n+1):
    c=a+b
    a=b
    b=c
    print(c)
fib(5)
```

```python
#palindrome
def is_palindrome(a):
  m=a
  num=0
  while m!=0:
    num=num*10+(m%10)
    m=m//10
  if(a==num):
    return True
  else:
    return False
print("Number 1312 is palindrome -",is_palindrome(1312))
```

```python
#count vowels in string
def count_vowels(a):
  cnt=0
  for i in range(0,len(a)):
    if(a[i]=='a'or a[i]=='e'or a[i]=='i'or a[i]=='o'or a[i]=='u'):
      cnt+=1
  return cnt
print("Number of vowels in SOUND is",count_vowels("soundaeu"))
```
